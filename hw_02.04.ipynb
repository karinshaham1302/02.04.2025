```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
sport_athletes = pd.read_csv("sport_athletes.csv")
df_sport_athletes = sport_athletes.copy()
display(df_sport_athletes.info())
display(df_sport_athletes.describe().T)
```

```python
# 1.

display("number of unique sport types:", df_sport_athletes['Sport Type'].nunique())

groups_sport_Type = df_sport_athletes.groupby('Sport Type').agg({
    'Weight(kg)': 'min',
    'Height(cm)': 'max',
    'Wins': 'max'}).rename(columns={'Weight(kg)': 'min-Weight', 'Height(cm)': 'max-Height', 'Wins':'max-Wins'})
groups_sport_Type
```

```python
# 2.

df_sport_athletes.groupby('Sport Type')['Height(cm)'].nlargest(2)
```

```python
# 3.

groups = df_sport_athletes.groupby(['Country', 'Sport Type'])
display(groups.size())
print()

# a.
display('the highest statistics:', df_sport_athletes.groupby(['Country', 'Sport Type']).max(numeric_only=True))
print()

# b.
df_basketball_usa = df_sport_athletes.loc[(df_sport_athletes['Country'] == 'USA') & (df_sport_athletes['Sport Type'] == 'Basketball')]
max_df = df_basketball_usa[['Height(cm)', 'Weight(kg)', 'Wins']].max(numeric_only=True)
display('the highest statistics in usa basketball:', max_df)
print()

# c.
# ?

# d.
df_tennis_football = df_sport_athletes[df_sport_athletes['Sport Type'].isin(['Tennis', 'Football'])]
df_max = df_tennis_football.groupby(['Sport Type']).max(numeric_only=True)
display('the highest statistics for tennis and football:', df_max)
print()

# e.
display('swaplevel:', df_sport_athletes.groupby(['Country', 'Sport Type']).max().swaplevel().sort_index(level=['Sport Type', 'Country']))

# f.
group_stat = df_sport_athletes.groupby('Sport Type').agg(
    min_wins=('Wins', 'min'),
    max_wins=('Wins', 'max'),
    mean_wins=('Wins', 'mean'),
    std_wins=('Wins', 'std'),
    min_height=('Height(cm)', 'min'),
    max_height=('Height(cm)', 'max'),
    mean_height=('Height(cm)', 'mean'),
    std_height=('Height(cm)', 'std')
).reset_index()
display('the statistics for the values: min, max, avg and std for the columns: wins and height:', group_stat)
```

```python
# 4.

new_rows = pd.DataFrame({
    'Country': ['Canada', 'USA'],
    'Sport Type': ['Hockey', 'Skiing'],
    'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'],
    'Height(cm)': [183, 172],
    'Weight(kg)': [84, 68],
    'Wins': [3, 4],
    'Olympic Date': ['01/02/2002', '12/02/2010']
})

df_sport_athletes = pd.concat([df_sport_athletes, new_rows], ignore_index=True)
df_sport_athletes
```

```python
# 5.

athletes_smoking = pd.read_csv("athletes_smoking.csv")
df_athletes_smoking = athletes_smoking.copy()
display(athletes_smoking.info())
display(athletes_smoking.describe().T)
print()

# a.
display('inner join:', pd.merge(df_sport_athletes, athletes_smoking, on='Athlete Name', how='inner',suffixes=('_df_sport_athletes', '_athletes_smoking')))
print()

# b.
display('outer join:', pd.merge(df_sport_athletes, athletes_smoking, on='Athlete Name', how='outer'))
```
